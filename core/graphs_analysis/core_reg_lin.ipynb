## Read Data

## Empty list for every earning sector

```python
list_result_lin_1 = []
list_fig_1 = []
list_mod_log_2 = []
list_fig_2 = []
list_value_boxcox = []
list_r_sqrt_lin = []
list_r_sqrt_log = []
```

## Loop for every analysis with the function apply_regression

Check the tools for function apply_regression

```python
list_earnings = ['agri_fish_hunt', 'construction', 'manufacturing', 'wholesale_trade',
       'retail_trade', 'transport_warehouse_utilities', 'transport_warehouse',
       'edu_serv', 'health_social', 'acc_food_serv', 'other_ser', 'pub_admin']
chemical_ = 'Trihalomethane_value'

for earning_ in list_earnings:
    result_lin_1, fig_1, mod_log_2,\
    fig_2, value_boxcox, r_sqrt_lin,\
    r_sqrt_log = apply_regression(earning=earning_, chemical=chemical_,
                                  df=df_core)
    list_result_lin_1 += [result_lin_1]
    list_fig_1 += [fig_1]
    list_mod_log_2 += [mod_log_2]
    list_fig_2 += [fig_2]
    list_value_boxcox += [value_boxcox]
    list_r_sqrt_lin += [r_sqrt_lin]
    list_r_sqrt_log += [r_sqrt_log]
```

## CREATE DF FOR EVERY EARNING WITH REGRESSION LIN AND LOG

```python
df_results = pd.DataFrame({
    'earnings':list_earnings,
    'result_lin_1':list_result_lin_1,
    'fig_1':list_fig_1,
    'mod_log_2':list_mod_log_2,
    'fig_2':list_fig_2,
    'value_boxcox':list_value_boxcox,
    'r_sqrt_lin':list_r_sqrt_lin,
    'r_sqrt_log':list_r_sqrt_log,
             })
df_results.set_index('earnings', inplace=True)
```

## Results
